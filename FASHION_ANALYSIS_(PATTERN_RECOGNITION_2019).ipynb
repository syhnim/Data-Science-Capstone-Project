Correlation Heatmap

```python
import seaborn as sns #seaborn library
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Separate numerical and non-numerical columns

#Get the list of column names
column_names = merged_data.columns

#Initialize empty lists to store numerical and non-numerical columns
numerical_columns = []
non_numerical_columns = []

#Iterate over each column
for column in column_names:
  #Check if the column contains numerical data
  if pd.api.types.is_numeric_dtype(merged_data[column]):
    numerical_columns.append(column)
  else:
    non_numerical_columns.append(column)


# Print the resulting column lists
print("Numerical Columns:")
print(numerical_columns)
print()
print("Non-Numerical Columns:")
print(non_numerical_columns)
```

```python
from scipy.stats import chi2_contingency

def cramers_v(x,y):
  confusion_matrix = pd.crosstab(x,y)
  chi2, _, _, _= chi2_contingency(confusion_matrix)
  n = confusion_matrix.sum().sum()
  phi2 = chi2/n
  r, k = confusion_matrix.shape
  phi2corr = max(0, phi2 - ((k-1) * (r-1)) / (n-1))
  rcorr = r - ((r-1) ** 2) / (n-1)
  kcorr = k - ((r-1) ** 2) / (n-1)
  return np.round(np.sqrt(phi2corr / min((kcorr-1), (rcorr - 1))), 4)
```

```python
# compute correlation heatmap
correlation_matrix = pd.DataFrame(index = non_numerical_columns, columns=non_numerical_columns, dtype=float)
for i in range(len(non_numerical_columns)):
  for j in range(i, len(non_numerical_columns)):
    column1 = merged_data[non_numerical_columns[i]]
    column2 = merged_data[non_numerical_columns[j]]
    correlation_matrix.loc[non_numerical_columns[i], non_numerical_columns[j]] = cramers_v(column1, column2)
    correlation_matrix.loc[non_numerical_columns[j], non_numerical_columns[i]] = correlation_matrix.loc[non_numerical_columns[i], non_numerical_columns[j]]

#Create a heatmap of the correlation matrix
plt.figure(figsize=(10,8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.title('Correlation matrix')
plt.colorbar()
tick_marks = np.arange(len(non_numerical_columns))
plt.xticks(tick_marks, non_numerical_columns, rotation=45)
plt.yticks(tick_marks, non_numerical_columns)

#Display the correlation values on the heatmap
for i in range(len(non_numerical_columns)):
  for j in range(len(non_numerical_columns)):
    text = plt.text(j,i,correlation_matrix.iloc[i,j],
                    ha="center",va="center",color="black", fontsize=8)

plt.tight_layout()
plt.show()
```

```python
from mlxtend.preprocessing import TransactionEncoder
from pymining import seqmining
```

```python
# Convert the dataset into a list of sequences format
sequences = []
for row in merged_data.iterrows():
    sequences.append(row[1].dropna().values.tolist())
```

```python
# Apply the PrefixSpan algorithm
frequent_patterns = seqmining.freq_seq_enum(sequences, min_support=10000)
```
