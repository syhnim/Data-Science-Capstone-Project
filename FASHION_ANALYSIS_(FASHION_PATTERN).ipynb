Correlation Heatmap

```python
#Separate numerical and non-numerical columns

#Get the list of column names
column_names = products.columns

#Initialize empty lists to store numerical and non-numerical columns
numerical_columns = []
non_numerical_columns = []

#Iterate over each column
for column in column_names:
  #Check if the column contains numerical data
  if pd.api.types.is_numeric_dtype(products[column]):
    numerical_columns.append(column)
  else:
    non_numerical_columns.append(column)


# Print the resulting column lists
print("Numerical Columns:")
print(numerical_columns)
print()
print("Non-Numerical Columns:")
print(non_numerical_columns)
```

```python
from scipy.stats import chi2_contingency

def cramers_v(x,y):
  confusion_matrix = pd.crosstab(x,y)
  chi2, _, _, _= chi2_contingency(confusion_matrix)
  n = confusion_matrix.sum().sum()
  phi2 = chi2/n
  r, k = confusion_matrix.shape
  phi2corr = max(0, phi2 - ((k-1) * (r-1)) / (n-1))
  rcorr = r - ((r-1) ** 2) / (n-1)
  kcorr = k - ((r-1) ** 2) / (n-1)
  return np.round(np.sqrt(phi2corr / min((kcorr-1), (rcorr - 1))), 4)
```

```python
# compute correlation heatmap
correlation_matrix = pd.DataFrame(index = non_numerical_columns, columns=non_numerical_columns, dtype=float)
for i in range(len(non_numerical_columns)):
  for j in range(i, len(non_numerical_columns)):
    column1 = products[non_numerical_columns[i]]
    column2 = products[non_numerical_columns[j]]
    correlation_matrix.loc[non_numerical_columns[i], non_numerical_columns[j]] = cramers_v(column1, column2)
    correlation_matrix.loc[non_numerical_columns[j], non_numerical_columns[i]] = correlation_matrix.loc[non_numerical_columns[i], non_numerical_columns[j]]

#Create a heatmap of the correlation matrix
plt.figure(figsize=(10,8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.title('Correlation matrix')
plt.colorbar()
tick_marks = np.arange(len(non_numerical_columns))
plt.xticks(tick_marks, non_numerical_columns, rotation=45)
plt.yticks(tick_marks, non_numerical_columns)

#Display the correlation values on the heatmap
for i in range(len(non_numerical_columns)):
  for j in range(len(non_numerical_columns)):
    text = plt.text(j,i,correlation_matrix.iloc[i,j],
                    ha="center",va="center",color="black", fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# Get the correlations with range < 1
correlations_less_than_1 = correlation_matrix.unstack().sort_values(ascending=False)
correlations_less_than_1 = correlations_less_than_1[correlations_less_than_1 < 1]

# Get the top 10 highest correlations
top_correlations = correlations_less_than_1[:10]

# Display the top 10 highest correlations
print("Top 10 Highest Correlations with Range < 1:")
print(top_correlations)
```

# **PART 4: MODELLING**

Find the pattern of fashion in H&M by using PrefixSpan Algorithm

```python
!pip install pymining
```

```python
from mlxtend.preprocessing import TransactionEncoder
from pymining import seqmining
```

```python
# Convert the dataset into a list of sequences format
sequences = []
for row in products.iterrows():
    sequences.append(row[1].dropna().values.tolist())
```

```python
# Apply the PrefixSpan algorithm
frequent_patterns = seqmining.freq_seq_enum(sequences, min_support=100)
```

# **PART 5: EVALUATION**

Finding pattern of the dataset

```python
# Print frequent patterns

max_patterns = 10

# Limit the number of frequent patterns to 10
limited_patterns = list(frequent_patterns)[:max_patterns]

# Print limited frequent patterns
print("Limited Frequent Patterns:")
for i, pattern in enumerate(limited_patterns):
    print(pattern)
    if i + 1 == max_patterns:
        break
```

```python
import matplotlib.pyplot as plt

max_patterns = 10

# Limit the number of frequent patterns to 10
limited_patterns = list(frequent_patterns)[:max_patterns]

# Extract pattern labels and frequencies
pattern_labels = [str(pattern) for pattern in limited_patterns]
pattern_frequencies = [1] * len(limited_patterns)  # Set a frequency of 1 for each pattern if using a set

if isinstance(frequent_patterns, dict):
    pattern_frequencies = [frequent_patterns[pattern] for pattern in limited_patterns]

# Create a bar chart
plt.bar(pattern_frequencies,pattern_labels)
plt.xlabel("Pattern")
plt.ylabel("Frequency")
plt.title("Frequent Patterns")

# Display the chart
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes and edges
for pattern in frequent_patterns:
    for i in range(len(pattern)-1):
        G.add_edge(pattern[i], pattern[i+1])

# Position nodes using circular layout
pos = nx.circular_layout(G)

# Draw nodes and edges
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)

# Set title
plt.title('Frequent Patterns')

# Display the diagram
plt.axis('off')
plt.show()
```
